# 简单理解一下sepia的操作过程

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

def sepia(input_img):
    sepia_filter = np.array([
        [0.393, 0.769, 0.189], 
        [0.349, 0.686, 0.168], 
        [0.272, 0.534, 0.131]
    ])
    sepia_img = input_img.dot(sepia_filter.T)
    sepia_img /= sepia_img.max()
    return sepia_img

# 读取图片
image_path = "image/687987.jpg"  # 替换为实际的图片路径
input_image = Image.open(image_path)
# input_image
```

```python
# 将图片转换为NumPy数组
input_array = np.array(input_image)
```

```python
input_array.shape, input_array.dtype
```

```python
# 应用sepia滤镜
output_array = sepia(input_array)

# 将处理后的图像转换回PIL图像对象
output_image = Image.fromarray((output_array * 255).astype(np.uint8))
# output_image
```

```python

# 显示原始图像和处理后的图像
fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(input_image)
axs[0].set_title("Original Image")
axs[0].axis("off")
axs[1].imshow(output_image)
axs[1].set_title("Sepia Filtered Image")
axs[1].axis("off")

plt.show()
```
